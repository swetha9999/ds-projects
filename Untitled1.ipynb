CLUSTERING UNIVARIATE,BIVARIATE AND MULTIVARIATE

```python
c1=KMeans(n_clusters=3)
```

```python
c1.fit(df[['Annual Income (k$)']])
```

```python
df['Income_clusters']=c1.labels_
```

```python
df
```

```python
df['Income_clusters'].value_counts()
```

```python
c1.inertia_
```

```python
inertia_Scores=[]
for i in range(1,11):
    Kmeans=KMeans(n_clusters=i)
    Kmeans.fit(df[['Annual Income (k$)']])
    inertia_Scores.append(Kmeans.inertia_)
```

```python
inertia_Scores
```

```python
plt.plot(range(1,11),inertia_Scores)
```

```python
df
```

```python
df.groupby("Income_clusters")[['Age','Annual Income (k$)','Spending Score (1-100)']].mean()
```

```python
inertia_Scores2=[]
for i in range(1,11):
    Kmeans=KMeans(n_clusters=i)
    Kmeans.fit(df[['Annual Income (k$)','Spending Score (1-100)']])
    inertia_Scores2.append(Kmeans.inertia_)
inertia_Scores2
```

```python
plt.plot(range(1,11),inertia_Scores2)
```

```python
c2=KMeans(n_clusters=5)
c2.fit(df[['Annual Income (k$)','Spending Score (1-100)']])
df['Spending and income cluster']=c2.labels_
```

```python
df
```

```python
centers=pd.DataFrame(c2.cluster_centers_)
centers.columns=['x','y']
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x=centers['x'],y=centers['y'],s=100,c="black",marker="*")
sns.scatterplot(data=df,x ="Annual Income (k$)",y ="Spending Score (1-100)",hue ="Spending and income cluster",palette ="tab10")
plt.savefig("c2.bivariate.png")
```

```python
pd.crosstab(df["Spending and income cluster"],df["Gender"])
```
